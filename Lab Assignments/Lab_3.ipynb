#Lab Tasks:

1. $\int \sin (a x) e^{-b x} d x$

2. $\int_{0}^{\infty} \frac{\sin ^{-1}(2 x)}{1+x^{2}} d x$

3. $\int_{0}^{2 \pi} \frac{1}{(a-\cos (x))^{2}+(b-\sin (x))^{2}} d x$

4. $\int_{0}^{\infty} \frac{\cos ^{-1}(2 x)}{1+x^{2}} d x$

5. $\int_{-1}^{2} x e^{6 x} d x$

#Solution:
#Task1

```python
x, a, b = smp.symbols("x, a, b")
fun = smp.sin(a*x)*smp.exp(-b*x)
smp.integrate(fun, x)
```

```python
smp.integrate(fun, x).simplify()
```

#Task2

```python
x = smp.symbols("x")
fun = (smp.sin(2*x)**-1) / (1 + x**2)
smp.integrate(fun, (x, 0, np.inf))
```

```python
x = smp.symbols("x")
fun = lambda x : (smp.sin(2*x)**-1) / (1 + x**2)
quad (fun, 0, np.inf) 
```

```python
quad (fun, 0, np.inf) [0]
```

#Task3

```python
x,a,b=smp.symbols("x,a,b")
fun =1/((a-smp.cos(x))**2)+((b-smp.sin(x))**2)
smp.integrate(fun,(x,0,2*smp.pi)).simplify()
```

#Task4

```python
x = smp.symbols("x")
fun = smp.cos(2*x)**-1 / (1+x**2)
smp.integrate(fun, (x, 0, np.inf))
```

```python
smp.integrate(fun, (x, 0, np.inf)).simplify()
```

```python
x = smp.symbols("x")
fun = lambda x : smp.cos(2*x)**-1 / (1+x**2)
quad (fun, 0, np.inf) 
```

```python
quad (fun, 0, np.inf) [0]
```
