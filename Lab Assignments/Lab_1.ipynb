## Variables

```python
a = 1       # integer
b = 1.1     # float or double
c = 1 + 2j  # complex number (a + bi)
d = "a"     # string
e = True    # boolean (True / False)
```

```python
print(a)
print(b)
print(c)
print(d)
print(e)
print(type(a))
```

##Type Conversion

```python
print(int(b))
print(float(a))
```

## Strings

```python
x = "Python"
print(x[0])
print(x[-1])
print(x[0:3])
print(x[0:3:2])
```

```python
print(len(x))
print(x.upper())
print(x.find("a"))
print(x.find("y"))
```

##Conditional Statements

```python
g=4

if g==0:
  print("Zero")
elif g>0:
  print("Positive")
else:
  print("Negative")
```

##Loops

###For Loop

```python
for n in range(1, 10): 
    print(n)
```

```python
for n in range(1, 10,2): 
    print(n)
```

```python
for n in range(10, 0,-1): 
    print(n)
```

###While Loop

```python
n=1
while n < 10: 
    print(n)
    n += 1   # n=n+1
```

##Functions and Taking input

```python
def increment(number, inc): 
   result=number + inc  
   return result
 
# Keyword arguments
print(increment(2, 3)) 
```

```python
#Taking an Input

inputnumber=int(input("Please enter a number: "))

print(increment(inputnumber, 1)) 
```

##Swapping Values

```python
x=10
y=22
x,y=y,x

print(x)
print(y)
```

##Lists (For more information click [here](https://www.w3schools.com/python/python_lists.asp))

```python
letters = ["a", "b", "c"] 

print(letters[0:3])   # "a", "b", "c"
print(letters[:3] )   # "a", "b", "c"
print(letters[0:])    # "a", "b", "c"
print(letters[:])     # "a", "b", "c"
print(letters[::2])   # "a", "c"
print(letters[::-1])  # "c", "b", "a" 

```

```python
letters.append("z")
print(letters)
```
